```python
# librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel, mutual_info_classif, RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# correlación (point-biserial = Pearson entre binaria y continua funciona)
corr = df.corr()  # matriz completa
plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title('Heatmap de correlaciones')
plt.show()

# correlación con el target
corr_target = corr['diagnóstico'].abs().sort_values(ascending=False)
print(corr_target)

# columnas con |corr| > 0.6 (excluye la propia variable objetivo)
strong_with_target = corr_target[corr_target.index != 'diagnóstico'][corr_target > 0.6].index.tolist()
print("Features con |corr|>0.6:", strong_with_target)
```

```python
#grafica comparativa entre perarson y no lineal
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(mi_series, label="MI series")
plt.plot(corr_target, label="Correlation target")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title("Comparación de MI vs Pearson")
plt.legend()
plt.show()
```

```python
# Alinear los índices para que ambos tengan las mismas variables y orden
common_idx = mi_series.index.intersection(corr_target.index)
mi_aligned = mi_series[common_idx]
corr_aligned = corr_target[common_idx]

plt.figure(figsize=(6,6))
plt.scatter(mi_aligned, corr_aligned, alpha=0.7)
plt.xlabel("MI series")
plt.ylabel("Correlation target")
plt.title("Relación entre MI y Correlación")
plt.show()
```

```python
#Mutual info detecta relaciones no lineales que Pearson puede ignorar.
#debido a que perason es el de -1 a 1 de la matriz de correlacion
X = df.drop(columns=['diagnóstico'])
y = df['diagnóstico']
mi = mutual_info_classif(X.fillna(X.mean()), y, discrete_features=False)
mi_series = pd.Series(mi, index=X.columns).sort_values(ascending=False)
print(mi_series.head(20))
```

```python
# ejemplo: L1 Logistic (regularización L1) para selección
scaler = StandardScaler()
Xs = scaler.fit_transform(X.fillna(X.mean()))

lr_l1 = LogisticRegression(penalty='l1', solver='saga', max_iter=5000, class_weight='balanced')
sel = SelectFromModel(lr_l1).fit(Xs, y)
selected_features = X.columns[sel.get_support()].tolist()
print("Seleccionadas por L1:", selected_features)

# ejemplo: importancia con RandomForest
rf = RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced')
rf.fit(X.fillna(X.mean()), y)
importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print(importances.head(20))
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=5000, class_weight='balanced'))
])

# usa solo features elegidas, por ejemplo selected_features
X_sel = df[selected_features].fillna(df[selected_features].mean())
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(pipeline, X_sel, y, cv=skf, scoring='roc_auc')
print("ROC AUC CV:", scores.mean(), scores)
```
